```python


ninteen80 = df[(df['Year'] > 1980) & (df['Year'] < 1989)]
ninteen90 = df[(df['Year'] > 1990) & (df['Year'] < 1999)]
twenty0 = df[(df['Year'] > 2000) & (df['Year'] < 2009)]
twenty10 = df[(df['Year'] > 2010) & (df['Year'] < 2020)]
```

```python
eighties = pd.DataFrame(ninteen80.groupby('Genre')['NA_Sales', 'EU_Sales','JP_Sales', 'Global_Sales'].mean())
eighties.reset_index(inplace=True)
eighties.rename(columns={'Sales': 'eighties'},inplace=True)


nineties = pd.DataFrame(ninteen90.groupby('Genre')['NA_Sales', 'EU_Sales','JP_Sales', 'Global_Sales'].mean())
nineties.reset_index(inplace=True)
nineties.rename(columns={'Sales': 'nineties'},inplace=True)

twenty = pd.DataFrame(twenty0.groupby('Genre')['NA_Sales', 'EU_Sales','JP_Sales', 'Global_Sales'].mean())
twenty.reset_index(inplace=True)
twenty.rename(columns={'Sales': '2000s'},inplace=True)


twenty10 = pd.DataFrame(twenty10.groupby('Genre')['NA_Sales', 'EU_Sales','JP_Sales', 'Global_Sales'].mean())
twenty10.reset_index(inplace=True)
twenty10.rename(columns={'Sales': '2010s'},inplace=True)

```

```python
###Mak a histogram for above and compare the results 

k = [eighties, nineties, twenty, twenty10]
for i in k: 
    fig = plt.figure()
    sns.histplot(data= i,x = 'Global_Sales', hue = 'Genre', multiple ='stack' )
    plt.show()
    
## Here we can see that simulation and sports game were unpopular world wide in the eighties and nineties
## Then sports started picking up in 2000s and 2010s probably because the popularity od sports teams
## Strategy games which started from thr nineties never went big as a matter of fact it got less popular over time
## Then platform stayed constant throughout time.
```

```python
##Create Dummy Variables
df_genre = pd.get_dummies(df['Genre'])
df = pd.concat([df, df_genre], axis=1).drop('Genre', axis=1)

df_platform = pd.get_dummies(df['Platform'])
df = pd.concat([df, df_platform], axis=1).drop('Platform', axis=1)

df_publisher = pd.get_dummies(df['Publisher'])
df = pd.concat([df, df_publisher], axis=1).drop('Publisher', axis=1)
```

```python
###Multiple Linear Regression Model
df = df.drop(columns = ['Name', 'EU_Sales', 'NA_Sales', 'Other_Sales', 'JP_Sales'])
X = df.drop(columns = 'Global_Sales')

y = df['Global_Sales']

from sklearn import linear_model
import statsmodels.api as sm
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
from sklearn.metrics import mean_squared_error as mse

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)
linreg.fit(X_train, y_train)
y_hat = linreg.predict(X_test)
y_hat_train = linreg.predict(X_train)
print('R_squared Score:', linreg.score(X, y))
    #Display errors
print('Root Mean Squared Error test:', np.sqrt(mse(y_test, y_hat)))
print('Root Mean Squared Error train:', np.sqrt(mse(y_train, y_hat_train)))
    #Compare predicted and actual values
print('Mean Predicted Selling Price:', y_hat.mean())
print('Mean Selling Price:', y_test.mean())





##Fernando Aguillar Medium
```
